```python
#Necessary Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
#Data Preprocessing
datos = pd.read_csv(r"data.csv")
print(datos)
```

```python
#Data Preprocessing
#Let's assign variables to the rows that we're going to use.
X = datos["Sales"]
Y = datos["TV"]
```

```python
#Data Preprocessing
#Let's try to plot it
plt.scatter(X,Y)
plt.show
```

```python
#Building the model
m = 0
c = 0

L = 0.0001
epochs = 1000
n = float(len(X))

for i in range(len(X)):
    Y_pred = m*X + c
    D_m = (-2/n) * sum(X * (Y - Y_pred))
    D_c = (-2/n) * sum(Y - Y_pred)
    m = m - L*D_m
    c = c - L*D_c
    
print(m, c)
```

```python
#Predictions Making
Y_pred = m*X + c

plt.scatter(X,Y)
plt.scatter(X,Y_pred)
plt.show
```

```python
#If we want a line to replace the optimal values, let's do it by
Y_pred = m*X + c

plt.scatter(X,Y)
plt.plot([min(X),max(X)], [min(Y_pred), max(Y_pred)], color = 'red')
plt.show()
```
